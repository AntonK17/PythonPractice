## Методы прямоугольника, трапеции и Симпсона
$$\int_0^1 \frac4{x^2+1} \,dx= \pi$$

### Метод трапеций

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def kernel(x):
    return 4/(1+x*x)

def getPlot(a,b,n):
    x = np.linspace(a,b,n)
    plt.plot(x,kernel(x)) #linewidth=3

def getGraph(method): 
    N=8
    precList=[0]*N
    resList=[0]*N
    plt.rcParams["figure.figsize"] = (15,15)
    for k in range(N):
        sumList = [[0]*2**k]*N
    for k in range(N):
        resList[k], precList[k], sumList[k] = method(0,1,2**k)
        print(sumList[k])
        getPlot(0,1,100)
        plt.title('Trapezoid method, Dim = {}'.format(2**k))
        plt.xlim([0, 1]) 
        plt.show()

    numbers=np.arange(0, N).tolist()
    splitDim = []
    for i in numbers:
        splitDim.append(2**i)    
    df = pd.DataFrame({'N': numbers,
                   'Split Dim': splitDim,
                   'Precision': precList,
                   'Results' : resList})
    df.set_index('N', inplace=True)
    print(df)
    return resList, precList
    
def trapezMethod(a,b,n):
    h=(b-a)/n
    x=a
    res=0
    listSum = []
    for k in range(1,n+1):
        xs = [x,x,x+h,x+h]
        ys = [0,kernel(x),kernel(x+h),0]
        plt.fill(xs,ys,'b',edgecolor='b',alpha=0.2)
        sumar=1/2*(kernel (x) + kernel (x+h))*h
        res+=sumar
        x=x+h
        listSum.append(sumar)
    precision = abs(res - np.pi)
    return res, precision, listSum

res1, precision1 = getGraph(trapezMethod)
```

### Метод Симпсона

```python
def coefGetter(x,h):
    a = np.array([[(x-h)**2, x-h , 1], 
                  [(x-h/2)**2, x-h/2, 1],
                  [x**2, x , 1]])
    b = np.array([kernel(x-h),kernel(x-h/2),kernel(x)])
    y = np.linalg.solve(a, b)
    return y

def simpsonMethod(a,b,n):   
    h=(b-a)/n
    x=a
    q=[]
    sumar=kernel(x)-kernel(b)
    listSum = []
    for k in range(1,n+1):
        x=x+h/2
        sumar+=4*kernel(x)
        x=x+h/2
        buf=2*kernel(x)
        sumar+=buf
        listSum.append(buf*h/6)
        q = coefGetter(x,h)
        xs = np.linspace(x-h, x, 10)
        ys = q[0]*xs**2 + q[1]*xs + q[2]
        plt.plot(xs, ys, linestyle = ':',
        linewidth = 6)
    sumar*=(h/6)
    precision = abs (sumar - np.pi)
    return sumar, precision, listSum

res2, precision2 = getGraph(simpsonMethod)
```

### Метод прямоугольников (средних)

```python
def midRectangle(a,b,n):
    h=(b-a)/n
    x=a
    res=0
    listSum = []
    for k in range(1,n+1):
        xs = [x,x,x+h,x+h]
        ys = [0,kernel(x+h/2),kernel(x+h/2),0]
        plt.fill(xs,ys,'b',edgecolor='b',alpha=0.2)
        sumar=(kernel(x+h/2))*h
        res+=sumar
        listSum.append(sumar)
        x=x+h
    precision = abs(res - np.pi)
    return res, precision, listSum

res3, precision3 = getGraph(midRectangle)
    
```

## Итоги

```python
N=8
splitDim=[]
for i in range(N):
    splitDim.append(2**i)    
df = pd.DataFrame({'Split Dim': splitDim,
                   'Prec Trap': precision1,
                   'Results Trap' : res1,
                   'Prec Simps':  precision2,
                   'Results Sipms': res2,
                   'Prec Rect':  precision3,
                   'Results Rect' : res3})
df.set_index('Split Dim', inplace=True)
pd.set_option('display.width', 100)
print(df)

fig, ax = plt.subplots()



ax.plot(splitDim, precision1, label = 'Trapezoid')
ax.plot(splitDim, precision2, label = 'Simpson')
ax.plot(splitDim, precision3, label = 'Rectangle')

ax.legend()

fig.set_figheight(5)
fig.set_figwidth(8)
plt.xlabel("Dimension", fontsize=14)
plt.ylabel("Precision", fontsize=14)
plt.xlim([0, 16]) 
plt.show()
```

Итоговая графика наглядно демонстрирует разницу скорости сходимости трех разных методов. Увеличивая сетку в 2 раза, методы прямоугольников и трапеции позволяют снизить невязку в 4 раза (квадратичная зависимость), а метод Симпсона является методом более высокого порядка точности – четвертого. При больших разбиениях из-за погрешности самих машинных вычислений и самого метода, невязка может перестать существенно изменяться.
