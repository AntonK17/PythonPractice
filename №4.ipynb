```python
from collections import defaultdict
import numpy as np
import pandas as pd
import multiprocessing as mp
#critDict = {'cpgs_in': ['cg00000029', 'cg00031443'],'gene_in': ['TMEM49'], 'cpgs_out': 'cg00002749','chr_in': 7}
#critDict = {'gene_out': ['TMEM49'], 'gene_in': ['PNPLA6'], 'cpgs_in': ['cg00000029', 'cg00031443']}
critDict = {'gene_out': float('nan'), 'chr_in': '7'}
testTable = [['cg00000029',16, 53468112,'RBL2','NM_005611','TSS1500','chr16:53468284-53469209','N_Shore','ClassA','chr16:53468284-53469209*N_Shore',5,0],
            ['cg00002749',17,57839538,'RBL2','NM_005611','TMEM49','NM_030938','Body','ClassC','NA','sdas','adgags',0,1,3]]

origNames = ['ID_REF','CHR','MAPINFO', 'Probe_SNPs', 'Probe_SNPs_10', 'UCSC_REFGENE_NAME', 'UCSC_REFGENE_ACCESSION','UCSC_REFGENE_GROUP','UCSC_CPG_ISLANDS_NAME','RELATION_TO_UCSC_CPG_ISLAND','Class','BOP','n.CpG','CROSS_R']

oldNames = ['ID_REF', 'CHR', 'UCSC_REFGENE_NAME', 'RELATION_TO_UCSC_CPG_ISLAND',
           'CROSS_R', 'Class', 'UCSC_REFGENE_GROUP']

df=pd.read_csv('C:/Users/Антонио/Documents/PythonProjects/cpgs_annotations_short (1).tsv', sep='\t')
lol = df.values.tolist()


class cpgs_annotation:
    def __init__ (self, table, origNames):
        self.table = table
        self.origNames = origNames
    def get_cpgs(self, critDict):
        self.critDict=critDict
        cpgsNames=[]
        indexes=[]
        newNames= ['cpgs' ,'chr' ,'gene','geotype','crossr','class','genepart']
        oldNames = ['ID_REF', 'CHR', 'UCSC_REFGENE_NAME', 'RELATION_TO_UCSC_CPG_ISLAND',
           'CROSS_R', 'Class', 'UCSC_REFGENE_GROUP']
        #Карта значений столбцов для 7 интересующих значений
        numNamesMap = {'cpgs' : 0,'chr' : 1,'gene' : 2,'geotype' : 3,'crossr' : 4,'class' : 5,'genepart' : 6}
        iterNames = iter(numNamesMap)
        for it in oldNames:
            numNamesMap[next(iterNames)]= self.origNames.index(it)
        
        temp={k: v for k, v in sorted(numNamesMap.items(), key=lambda item: item[1])}
        numNamesMap=temp

        aproveMap = defaultdict(list)
        disaproveMap  = defaultdict(list)
        
        for iterNames, iterVals in self.critDict.items():            
            key = iterNames.split('_')
            if key[0] not in newNames:
                continue
            if isinstance(iterVals, float):
                continue
            if len(key) != 2:
                continue
            critName, critFlag = key          
            if (critFlag == 'in'):
                if (isinstance(iterVals,list)):
                    aproveMap[critName].extend(iterVals)
                elif( (isinstance(iterVals,int) or (isinstance(iterVals,float)))):
                    if not np.isnan(iterVals):
                        aproveMap[critName].append(str(iterVals))
                else:
                    aproveMap[critName].append(str(iterVals))
            elif (critFlag == 'out'):
                if (isinstance(iterVals,list)):
                    disaproveMap[critName].extend(iterVals)
                elif((isinstance(iterVals,int) or (isinstance(iterVals,float)))):
                    if not np.isnan(iterVals):
                        disaproveMap[critName].append(str(iterVals))
                else:
                    disaproveMap[critName].append(str(iterVals))
        print(numNamesMap)            
        print(aproveMap)
        print(disaproveMap)
        
        
        for i in range(len(self.table)):
            flag=False
            for iterNames, iterVals in aproveMap.items():
                temp = str(self.table[i][numNamesMap[iterNames]]).split(sep=';')
                for it in temp:
                    if it in iterVals:
                        flag=True
                    else:
                        flag=False
                if (flag==False):
                    break
            if flag==True:
                for iterNames, iterVals in disaproveMap.items():
                    temp = str(self.table[i][numNamesMap[iterNames]]).split(sep=';')
                    for it in temp:
                        if it in iterVals:
                            flag=False
                    if (flag==False):
                        break
            if flag==True:
                indexes.append(i)
                cpgsNames.append(self.table[i][numNamesMap['cpgs']])
                          
        return cpgsNames, indexes
```
